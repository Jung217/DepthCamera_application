```python
import matplotlib.pyplot as plt
import numpy as np

from skimage.io import imread, imshow
from skimage.color import rgb2hsv, rgb2gray
from skimage.exposure import histogram, cumulative_distribution
from skimage.filters import threshold_otsu
```

```python
#test = imread('test.png')
test = imread("/content/drive/MyDrive/NTUST/DATA/T2/img_1720488725.png")
imshow(test)
```

```python
test_gray = rgb2gray(test)
imshow(test_gray, cmap='gray')
```

```python
th = 0.4
test_gray_bw = test_gray < th
imshow(test_gray_bw, cmap='gray')
```

```python
test_hsv = rgb2hsv(test)

fig, ax = plt.subplots(1, 3, figsize=(12,4))

ax[0].imshow(test_hsv[:,:,0], cmap='gray')
ax[0].set_title('Hue')

ax[1].imshow(test_hsv[:,:,1], cmap='gray')
ax[1].set_title('Saturation')

ax[2].imshow(test_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```
